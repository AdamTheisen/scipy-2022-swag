```python
import yaml
with open('data.yml') as f:
    data = yaml.safe_load(f.read())
data
```

```python
def f(color, power, desc, short, name):
    desc = desc.strip()
    nl = len(desc.split('\n'))
    print(nl)
    desc + (5-nl)*'\n'
    import matplotlib.pyplot as plt
    import matplotlib.image as mpimg
    img = mpimg.imread(f'createcard-{color}.jpg')
    logo = mpimg.imread(f'logos/{name.lower()}.png')
    py = mpimg.imread('python.png')
    fig, ax = plt.subplots()
    fig.set_figheight(12)
    #fig.set_figwidth(20)
    ax.imshow(img)
    ax.text(50, 52, name, fontsize = 18, fontfamily='cursive')
    ax.text(39, 334, short, fontsize = 18,fontfamily='cursive')
    ax.text(39, 500, desc, fontsize = 16, fontfamily='cursive')
    ax.text(49, 532, 'trade me with other attendees, and come see NumFOCUS Booth', fontsize = 10, 
            fontfamily='cursive', color=('white' if color=='black' else 'black')
    )
    

    for n in range(power):

        pyax = fig.add_axes([0.78-n*0.040,0.705,0.033,0.045], anchor='C', zorder=1)
        pyax.imshow(py)
        pyax.patch.set_alpha(0)
        pyax.axis('off')  
    
    newax = fig.add_axes([0.20,0.48,0.62,0.22], anchor='C', zorder=1)
    newax.imshow(logo)
    newax.patch.set_alpha(0.0)
    newax.axis('off')
    ax.axis('off')
    fig.savefig(f'{name}-card.png', dpi=300)


    
for kw in data[-5:]:
    f(**kw)

```
